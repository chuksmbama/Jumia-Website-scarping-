JUMIA WEBSITE SRAPING 

```python
# Importing Libaries 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Jumia URL
base_url = "https://www.jumia.com.ng/phones-tablets/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36"}
```

```python
# Empty lists to store scraped data
product_names = []
product_prices = []
product_links = []
```

```python
# Scraping first 5 pages
for page in range(1, 6):
    print(f"Scraping Jumia Page {page}...")
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")
    
    # Find all product cards
    products = soup.find_all("div", class_="itm col")
    
    for item in products:
        name = item.find("div", class_="name")
        price = item.find("div", class_="prc")
        link = item.find("a", href=True)
        
        if name and price and link:
            product_names.append(name.text.strip())
            product_prices.append(price.text.strip().replace("₦", "").replace(",", ""))
            product_links.append("https://www.jumia.com.ng" + link["href"])
    
    time.sleep(2)  
```

```python
# Create DataFrame
jumia_df = pd.DataFrame({
    "Product Name": product_names,
    "Price (₦)": product_prices,
    "Product Link": product_links
})

jumia_df.to_csv("jumia_products.csv", index=False)
print("Jumia data saved to jumia_products.csv")
jumia_df.head()
```
